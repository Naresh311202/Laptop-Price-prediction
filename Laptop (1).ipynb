 **Laptop Prices Dataset**

This dataset is a collection of features related to various laptops, such as brand, processor type, RAM, storage capacity, and other specifications. The dataset also includes the corresponding prices of these laptops. This dataset can be used for regression analysis to predict the prices of laptops based on their features.

**1.Problem Definition**

Building regression models to predict the prices of laptops based on various features.

**2.Data.**

The dataset was downloaded in kaggle: https://www.kaggle.com/datasets/anubhavgoyal10/laptop-prices-dataset

Convert the object dtypes into string dtypes

```python
for label,content in laptop.items():
  if pd.api.types.is_object_dtype(content):
    laptop[label]=content.astype("string")
```

Convert the string into category

```python
for label,content in laptop.items():
  if pd.api.types.is_string_dtype(content):
    laptop[label]=content.astype("category").cat.as_ordered()
```

```python
laptop.info()
```

```python
pd.Categorical(laptop["rating"]).codes
```

Convert into number

```python
for label,content in laptop.items():
  if not pd.api.types.is_numeric_dtype(content):
    laptop[label]=pd.Categorical(content).codes+1
```

**Import the Evaluation Metrics**

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def evaluation_metrics(y_true,y_preds):
  mae=mean_absolute_error(y_true,y_preds)
  mqe=mean_squared_error(y_true,y_preds)
  r2=r2_score(y_true,y_preds)
  metrics={"r2_score":round(r2,2),
           "mean_absoluate_error":round(mae,2),
           "mean_squared_error":round(mqe,2)}
  print(f"the r2 score of the model:{r2*100:.2f}%")
  print(f"the mean absolute error of the model:{mae}")
  print(f"the mean squared error of the model :{mqe}")
  return metrics
```

```python
baseline_model_score=evaluation_metrics(y_test,clf_pred)
```

```python
randomized_score=evaluation_metrics(y_test,rand_preds)
```

```python
gridsearchcv_score=evaluation_metrics(y_test,grid_pred)
```
